```python
# functions_for_lab
def initialize_clenaing(df):
    # Make column names in lower case, and change st for state, replace whitespace for _
    df.columns = df.columns.str.lower()
    df = df.rename(columns={"st": "state"})
    df.columns = df.columns.str.replace(" ", "_")

    # Clean gender data
    df.loc[df["gender"].str.lower().str.startswith("f", na = False), "GENDER"] = "F"
    df.loc[df["gender"].str.lower().str.startswith("m", na = False), "GENDER"] = "M"

    # Replace abvrevations in state column by their full name
    df["state"] = df["state"].replace({"AZ": "Arizona", "Cali": "California", "WA": "Washington"})
    df["education"] = df["education"].replace({"Bachelors": "Bachelor"})

    # Remove % in customer_lifetime_value
        # type(df["customer_lifetime_value"][5]) # the data is a string
    df["customer_lifetime_value"] = df["customer_lifetime_value"].str.replace("%", "")

    df["vehicle_class"] = df["vehicle_class"].replace({"Sports Car": "Luxury",
                                                    "Luxury SUV": "Luxury", 
                                                    "Luxury Car": "Luxury"})

# Handling Customer lifetime value to numeric
    df["customer_lifetime_value"] = df["customer_lifetime_value"].astype(float)

    # Handling number_of_open_complaints
    df["number_of_open_complaints"] = df["number_of_open_complaints"].str.replace(r'\d{1}\/(\d{1})\/\d{2}', r'\1', regex=True)
    return df
```

```python
# Your code goes here
df_4 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv").iloc[ :, 1:]

df_4.head()
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python

 # Slice to have the relevant columns
df_5 = df_4[["sales_channel", "monthly_premium_auto", "months_since_policy_inception", "customer_lifetime_value","gender", "education"]]
 # Calculate the total revenue for each customer
df_5["lifetime_revenue"] = df_5["monthly_premium_auto"] * df_5["months_since_policy_inception"]

df_5.head(10)

```

```python
df_5.pivot_table(
    values = "lifetime_revenue",
    index = "sales_channel",
    aggfunc = "sum"
    ).round(2).sort_values("sales_channel")
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
df_5.head(2)
```

```python
df_5.pivot_table(
    values = "customer_lifetime_value",
    index = ["gender", "education"],
    aggfunc = "mean"
    ).round(2).sort_values(["gender", "customer_lifetime_value"], ascending=False)

# Based on the data, most of the company revenue comes from individuals with lower education level.
# This is likely because there are more individuals on that group.
```

You work at the customer service department and you want to know which months had the highest number of complaints by policy type category. Create a summary table showing the number of complaints by policy type and month.
Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
df_6 = df_4[["months_since_last_claim", "policy_type"]]
df_6["months_since_last_claim"] = df_6["months_since_last_claim"].astype(int)

from datetime import datetime
from dateutil.relativedelta import relativedelta



datetime.now().date()-relativedelta(months=3)
df_6["date_last_claim"] = df_6["months_since_last_claim"].apply((lambda x: datetime.now().date() - relativedelta(months = x)))

df_6["month_last_claim"] = df_6["date_last_claim"].apply(lambda x: x.month)


```

```python
df_6

df_most_claims = df_6.pivot_table(
    values = "policy_type",
    index = "month_last_claim",
    aggfunc = "count"
)

df_most_claims
# The month with most claims is 7 (July)
```
